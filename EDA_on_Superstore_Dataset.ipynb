**Drop Unnecessary Columns**

```python
data = data.drop(['Row ID', 'Order ID','Customer ID','Postal Code'],axis=1)
```

```python
data.columns
```

**Hypothesis 1: Technology products have the hightest profit margin compared to other product categories.**

```python
data.columns
```

```python
cat_profit = data.groupby('Category')['Profit'].sum()
cat_profit.plot(kind='bar')
plt.title("Profit by Category")
plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.show()

# Conclusion : The hypothesis is supported as the technology products have the highest profit margin of the three category.
```

**Hypothesis 2: The East region has the highest sales as compared to the other regions**


```python
data.columns
```

```python
reg_sales = data.groupby('Region')['Sales'].sum().sort_index(ascending=True)
reg_sales.plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.plot()

# Conclusion : The Hypothesis is not supported as the Central region has the highest sales.
```

**Hypothesis 3 : Sales are higher during the certain months of the year.**

```python
data.columns
```

```python
data['Order_Month'] = pd.DatetimeIndex(data['Order Date']).month
Sales_month = data.groupby('Order_Month')['Sales'].sum()
Sales_month.plot(kind='line')
plt.title("Sales by Month")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()

# Conclusion: The Hypothesis is supported as the sales are higher during the certain months of the year.
```

**Hypothesis 4 :Order with the same day shipping have the lowest rate of returnedf products.**

```python
data.columns
```

```python
total_ord_by_shipping_mode = data.groupby('Ship Mode').size()

returned_order_by_shipping_mode = data[data['Profit']<0].groupby('Ship Mode').size()

returned_per_shipping_mode= (returned_order_by_shipping_mode/total_ord_by_shipping_mode)*100
print(returned_per_shipping_mode)

Returned_per_shipping.plot(kind='bar')
plt.title("Order returned on Same Day Shipping")
plt.xlabel(" Shipping Mode ")
plt.ylabel("Returned Percentage")
plt.show()

# Conclusion: The Hypothesis is supported as order with the same day shipping have the lowest rate of returnedf products.
```

**Hypothesis 5:The companies profit is more on weekdays then weekends**

```python
  data.columns
```

```python
data['Order_Day']=pd.DatetimeIndex(data['Order Date']).day_name()

Day_profit = data.groupby('Order_Day')['Profit'].sum()
Day_profit.plot(kind ='bar')
plt.title("Profit Day Wise")
plt.xlabel(" Day of the Week")
plt.ylabel("Profit")
plt.show()

# Conclusion: The Hypothesis is supported as the companies profit is more on weekdays then weekends
```
